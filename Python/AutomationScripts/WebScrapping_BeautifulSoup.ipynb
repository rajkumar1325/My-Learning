```python
!pip install requests
import requests
from bs4 import BeautifulSoup
```

```python
# Target website URL
url = 'http://fangraphs.com/leaders/major-league'


# Fetch the webpage Content
try:
  response = requests.get(url)
  response.raise_for_status() #check for HTTP error

except requests.exceptions.RequestException as e:
    print(f"Error fetching the webpage {e}")

else:
  # parse the html content
  soup = BeautifulSoup(response.content, 'html.parser')

# now: soup holds a structured representation of the webpage, ready for exploration
```

finding all headings inside the website

```python
headings = soup.find_all('h1')
for head in headings:
  print(head.text.strip()) #strip--> removes extra whitespaces
```

Finding all links inside the website

```python
links = soup.find_all('a')
for link in links:
  href = link.get('href')
  if href:    #make sure the link has a valid 'href' attribute.
    print(href)
```

## Example
#### Sports data analyst

```python
url = 'https://www.fangraphs.com/leaders/major-league'

try:
  response = requests.get(url)
  response.raise_for_status()

except requests.exceptions.RequestException as e:
  print(f"error fetching the webpage {e}")

else:
  # parse the html content
  soup = BeautifulSoup(response.content, 'html.parser')


sports = soup.find_all('div', class_='leaders-major_leaders-major__table__hcmbm')


for item in sports:
  # player name
  player = item.find('td', class_='align-left fixed').text.strip()

  # number of games played by player
  game = item.find('td', class_='align-right').text.strip()

  print(f"player {player} and game played: {game}")
```
